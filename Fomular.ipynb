```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
blk = 1
whe = 0
```

<br> 1/2 Tex <= Ts <= Tex
<br> Tex = nTr (n<10)
<br> Tr = T row = 1
<br> 3Ts > Tex + Tr

```python
Tr=1
Tex= 3
Ts=1
```

```python
arr= np.arange(0, 15,1)
for i in range(15):
    T= i*Tr + Tex
    if(math.floor(T/Ts)%2==0):
        m=T- math.floor(T/Ts)*Ts
        if(Tex-m>Ts):
            n=Ts
            k=Tex-m
            m+=k-Ts
        else:
            n=Tex-m
        arr[i]= m*whe + n*blk;
    if (math.floor(T/Ts)%2==1):
        n=T- math.floor(T/Ts)*Ts
        if(Tex-n>Ts):
            m=Ts
            k=Tex-n
            n+=k-Ts
        else:
            m=Tex-n
        arr[i]= m*whe + n*blk;
    print("hang",i, T,  arr[i])
```

```python
plt.scatter( x=np.arange(0,15,1) ,y= arr)
plt.show()
```
